# Unemployment rates in 1980 to 1990

```python
sns.lineplot(data=df[(df.date>='1/1/1980')&(df.date<='12/1/1984')], x='date', y='unrate_women')
```

Unemployment rate in women peaked between 1983 to 1984

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df[(df.date>='01/1/1982')&(df.date<='01/1/1984')], x='date', y='unrate_women', color='pink')

plt.xticks(rotation=45);
```

Most jobs were lost in mid year (April 1982) indicating that women were first to let go

```python
plt.figure(figsize=(10,5))
sns.lineplot(data=df[(df.date>='1/1/1982')&(df.date<='1/1/1984')], x='date', y='unrate_men')
plt.xticks(rotation=45);
```

While most men lost their jobs starting July of 1982

Unemployment rates increased in 1982 for both men and women while men (unemployment rate=12.5) seem to be more affected than women ( unemployment rate=10.5)

```python

plt.figure(figsize=(30,10))
sns.lineplot(data=df[(df.date>='1/1/1981')&(df.date<='1/1/1984')],x="date",y="unrate_16_to_17",label="unrate_16_to_17")
sns.lineplot(data=df[(df.date>='1/1/1981')&(df.date<='1/1/1984')],x="date",y="unrate_18_to_19",label="unrate_18_to_19")
sns.lineplot(data=df[(df.date>='1/1/1981')&(df.date<='1/1/1984')],x="date",y="unrate_20_to_24",label="unrate_20_to_24")
sns.lineplot(data=df[(df.date>='1/1/1981')&(df.date<='1/1/1984')],x="date",y="unrate_25_to_34",label="unrate_25_to_34")
sns.lineplot(data=df[(df.date>='1/1/1981')&(df.date<='1/1/1984')],x="date",y="unrate_35_to_44",label="unrate_35_to_44")
sns.lineplot(data=df[(df.date>='1/1/1981')&(df.date<='1/1/1984')],x="date",y="unrate_45_to_54",label="unrate_45_to_54")
sns.lineplot(data=df[(df.date>='1/1/1981')&(df.date<='1/1/1984')],x="date",y="unrate_55_over",label="unrate_55_over")
```

The peaks after 1983 in young people points to the fact that more younger people ( below 19) lost their jobs in comparison to people above 20

# Analysing unemployment levels in the recession of 2010

```python
recession12=df[(df.date>='01/01/2009')&(df.date<='12/30/2014')]
recession12.tail()
```

```python
sns.lineplot(data=recession12, x='date',y='unrate')
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=recession12, x='date', y='unrate_men', label='men');

plt.legend;
```

```python
recession12['unrate_men'].describe()
```

From 2009 to 2015 the average unemployment in men stood at 8.7% with unemployment peaking in late 2009

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=recession12, x='date', y='unrate_women', label='women', color='pink');
```

```python
recession12['unrate_women'].describe()
```

From 2009 to 2015 the average unemployment in women stood at 7.7% with unemployment peaking after 2011

Men were clearly more affected in the recession than women by 3%. While both the above graphs points to the fact the economy was very unstable in 2012 and even after 2013 with varying levels of unemployment

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=recession12,x="date",y="unrate_16_to_17",label="unrate_16_to_17")
sns.lineplot(data=recession12,x="date",y="unrate_18_to_19",label="unrate_18_to_19")
sns.lineplot(data=recession12,x="date",y="unrate_20_to_24",label="unrate_20_to_24")
sns.lineplot(data=recession12,x="date",y="unrate_25_to_34",label="unrate_25_to_34")
sns.lineplot(data=recession12,x="date",y="unrate_35_to_44",label="unrate_35_to_44")
sns.lineplot(data=recession12,x="date",y="unrate_45_to_54",label="unrate_45_to_54")
sns.lineplot(data=recession12,x="date",y="unrate_55_over",label="unrate_55_over")
```

Unemployment levels touched skyhigh in 2010 with more than 35% eligible youth unemployed!

# Effect of covid on unemployment:

```python
covid_data=df[(df.date>='01/01/2020')]
covid_data.head()
```

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=covid_data,x="date", y='unrate_men', label="men")

plt.legend
```

```python
covid_data['unrate_men'].describe()
```

Most men lost their jobs in March of 2020 with unemployment rates peaking at 13.3

```python
plt.figure(figsize=(15,5))
sns.lineplot(data=covid_data, x="date", y="unrate_women", color='pink', label='women')
plt.legend
```

```python
covid_data['unrate_women'].describe()
```

Most women lost their jobs in March of 2020 with unemployment rates peaking at 15.7

From the above data it is evident that women were most affected by the pandemic

```python
plt.figure(figsize=(15,10))
sns.lineplot(data=covid_data,x="date",y="unrate_16_to_17",label="unrate_16_to_17")
sns.lineplot(data=covid_data,x="date",y="unrate_18_to_19",label="unrate_18_to_19")
sns.lineplot(data=covid_data,x="date",y="unrate_20_to_24",label="unrate_20_to_24")
sns.lineplot(data=covid_data,x="date",y="unrate_25_to_34",label="unrate_25_to_34")
sns.lineplot(data=covid_data,x="date",y="unrate_35_to_44",label="unrate_35_to_44")
sns.lineplot(data=covid_data,x="date",y="unrate_45_to_54",label="unrate_45_to_54")
sns.lineplot(data=covid_data,x="date",y="unrate_55_over",label="unrate_55_over")
plt.ylabel("Unemployment rate")
```

While We had always seen higher peaks of unemployment in youth but amid the pandemic unemployment in people aged 20 and above rose too!
